```python
# create variable
v_num = 1190
variables = [pulp.LpVariable('X_%d'%i, upBound = 1, lowBound = 0, cat = pulp.LpInteger) for i in range(0, v_num)]

print(variables)

# objective function
c = df_c[0].tolist()
objective = sum([c[i] * variables[i] for i in range(0, v_num)])

# subjective function
constraints = []
b = df_b[0].tolist()
for i in range(133):
    A = df_A[i].tolist()
    constraints.append(sum([A[j] * variable[j] for j in range(0, v_num)]) <= b[i])

print(len(constraints))
```

```python
def solve_integer_LP_max(objective, subjective):
    # building problem
    problem = pulp.LpProblem('Integer Programming', pulp.LpMaximize)
    problem += objective
    for constraint in subjective:
        problem += constraint
    # solving problem
    solution = problem.solve()
    if solution != 1:
        print("Error in solve_integer_LP_max")
        return None
    else:
        # return a list of solutions, 1 if we select, 0 otherwise
        return [v.varValue.real for v in problem.variables()]
```

```python
res = solve_integer_LP_max(objective, constraints)
print(res)
```
